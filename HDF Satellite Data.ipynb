```python
import pandas as pd
from pyhdf.SD import SD, SDC
```

```python
file = SD('./MOD05_L2.A2020113.1630.061.hdf', SDC.READ)
datasets_dic = file.datasets()
```

```python
datasets_dic
```

```python
water_vapour = file.select('Water_Vapor_Infrared')
lat = file.select('Latitude')
lng = file.select('Longitude')
water = water_vapour.get()
latitude = lat.get()
longitude = lng.get()
water_data = pd.DataFrame(water)
lat_dat = pd.DataFrame(latitude)
lng_data = pd.DataFrame(longitude)
```

```python
lat_dat
```

```python
w,st,lt,lg = [],[],[],[]
for i in range(water_data.shape[0]):
    for j in range(water_data.shape[1]):
        try:
            w.append(water_data[i][j])
            lt.append(lat_dat[i][j])
            lg.append(lng_data[i][j])
        except:
            pass
print("Creating DataFrame")
df = pd.DataFrame()
df['lat'] = lt
df['lng'] = lg
df['water_vapour'] = w
```
